```python
# Calculate paper that have abstract and references for splitting text
# pypdf (jadi tidak menggunakan yg ini)
abstract = 0
references = 0
for file in files_new:
    new_name_file = file.replace('.pdf', '')
    
    with open('/Users/williamnehemia/Documents/Skripsi/TugasAkhir/DataPaperExtracted/' + new_name_file, 'r') as fileNow:
        content = fileNow.read()
        content = content.lower()
        if 'abstract' in content:
            abstract += 1
#         else:
#             print("no abstract", new_name_file)
        if 'references' in content:
            references += 1
#         else:
#             print("no references", new_name_file)
print('abstract', abstract)
print('references', references )
```

```python
# Calculate paper that have abstract and references for splitting text
# pymupdf (pakai yg ini)
abstract = 0
references = 0
for file in files_new:
    new_name_file = file.replace('.pdf', '')
    abs_flag = False
    ref_flag = False
    with open('/Users/williamnehemia/Documents/Skripsi/TugasAkhir/DataPaperExtractPyMuPDF/' + new_name_file, 'r') as fileNow:
        content = fileNow.read()
        content = content.lower()
        lines = content.splitlines()
        for line in lines:
            line = line.replace(" ", "")
            if ('abstract' in line) and (abs_flag == False) :
                abstract += 1
                abs_flag = True
            #else:
                #print("no abstract", new_name_file)
            if ('references' in line) and (ref_flag == False):
                references += 1
                ref_flag = True
            #else:
                #print("no references", new_name_file)
#         if abs_flag == False:
#             print("no abstract", new_name_file)
#         if ref_flag == False:
#             print("no references", new_name_file)
        if (abs_flag == True) and (ref_flag == True):
            source_file = '/Users/williamnehemia/Documents/Skripsi/TugasAkhir/DataPaperExtractPyMuPDF/' + new_name_file
            destination_folder = '/Users/williamnehemia/Documents/Skripsi/TugasAkhir/DataPaperExtractedWillBeUsed'
            shutil.copy(source_file, destination_folder)
        
print('abstract', abstract)
print('references', references )
```

```python
directory_path = '/Users/williamnehemia/Documents/Skripsi/TugasAkhir/DataPaperExtractedWillBeUsed'

files = os.listdir(directory_path)
files_new = []

min_after_abstract = []
delimiter_abs = "abstract"
delimiter_ref = "references"
for file in files:
    files_new.append(file)
    
for file in files_new:
    abs_flag = False
    ref_flag = False
    min_after_abstract_curr = sys.maxsize
    before_abstract_text = ""
    abstract_references_text = ""
    with open('/Users/williamnehemia/Documents/Skripsi/TugasAkhir/DataPaperExtractedWillBeUsed/' + file, 'r') as fileNow:
        content = fileNow.read()
        content = content.lower()
        lines = content.splitlines()
        
        for line in lines:
            line_curr = line.replace(" ", "")
            
            if ('abstract' in line_curr) and (abs_flag == False) : #checkpoint sudah mencapai abstract
                abstract += 1
                abs_flag = True
            elif abs_flag == True: #sedang di baris setelah abstract
                abstract_references_text += line + '\n'
                splited_line = line.split(" ")
                min_curr = len(splited_line)
                if (min_curr < min_after_abstract_curr):
                    print('---')
                    print(line)
                    print('---')
                    min_after_abstract_curr = min_curr
                if ('references' in line_curr) and (ref_flag == False):
                    break
            else: # sedang di baris sebelum abstract
                before_abstract_text += line +'\n'
                
        min_after_abstract.append(min_after_abstract_curr)
#     print('----', file, '----')
#     print(before_abstract_text)
#     print('---- abstract---')
#     print(abstract_references_text)
print(min_after_abstract)
            
        
```
